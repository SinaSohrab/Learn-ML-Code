```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# date
x = np.array([1, 2, 3, 4]).reshape(-1, 1)
y = np.array([1, 3, 5, 7])
```

```python
x
```

```python
# plot
plt.plot(x, y, "or")
plt.show()
```

```python
# import model
from sklearn.linear_model import LinearRegression
```

```python
# model
model = LinearRegression()
model.fit(x, y)
```

```python
plt.plot(x, y, "or")
plt.plot(x, model.predict(x))
plt.show()
```

```python
#َAsk questions tn the model
result = model.predict([[5]])
print(result)
```
